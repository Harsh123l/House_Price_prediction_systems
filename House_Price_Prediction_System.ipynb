Importing Dependency

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn.datasets
from sklearn.model_selection import train_test_split
from xgboost import XGBRegressor
from sklearn import metrics
```

Importing Boston House Price Data


```python
house_price_Dataset = sklearn.datasets.fetch_california_housing()
```

```python
print(house_price_Dataset)
```

```python
#loading the data into pandas dataframe
house_price_system_dataframes = pd.DataFrame(house_price_Dataset.data,columns=house_price_Dataset.feature_names)
```

```python
house_price_system_dataframes.head()
```

```python
house_price_system_dataframes['Prices'] = house_price_Dataset.target
```

```python
house_price_system_dataframes.head()
```

```python
#checking number of rows and columns
house_price_system_dataframes.shape
```

```python
#checking the null values or missing val from set
house_price_system_dataframes.isnull().sum()
```

```python
house_price_system_dataframes.describe()
```

understanding the correlation between various features in dataset


```python
correlation  = house_price_system_dataframes.corr()
```

```python
plt.figure(figsize=(10,10))
sns.heatmap(correlation, cbar=True, square=True, fmt ='.1f', annot=True, annot_kws= {'size':8}, cmap= 'Blues')
```

Splitting the data and target

```python
y = house_price_system_dataframes['Prices']
x = house_price_system_dataframes.drop(['Prices'] , axis=1)
```

```python
print(x)
print(y)
```

```python
x_train, x_test, y_train, y_test =  train_test_split(x,y,test_size=0.2,random_state= 2)
```

```python
print(x.shape,x_train.shape,x_test.shape)
```

Model training of the data

```python
#using XGBoost regression loading the model
model = XGBRegressor()
```

```python
model.fit(x_train,y_train)
```

**Evaluating the prediction on training**

```python
data_prediction = model.predict(x_train)
```

```python
print(data_prediction)
```

Now we will compare with actual values here we consider
**R squared error** & **MEAN Absolute Error** whereas in **SVM** we used **Accuraccy** for understanding


```python
# R squared error
score_1 = metrics.r2_score(y_train,data_prediction)
# Mean absolute values
score_2 = metrics.mean_absolute_error(y_train,data_prediction)

print("R squared error : ", score_1)
print("Mean Asolute error : ", score_2)
```

**Visualizing the actual prices with predicted prices using scatter plot**

```python
plt.scatter(y_train,data_prediction)
plt.xlabel("Actual Prices")
plt.ylabel("Predicted Prices")
plt.title("Actual Prices VS Predicted Prices")
plt.show()
```

```python
# now we are predicting for test data
test_data_prediction = model.predict(x_test)
```

```python
print(test_data_prediction)
```

Doing same for test data and then comparing

```python
# R squared error
score_3 = metrics.r2_score(y_test,test_data_prediction)
# Mean absolute values
score_4 = metrics.mean_absolute_error(y_test,test_data_prediction)

print("R squared error : ", score_3)
print("Mean Asolute error : ", score_4)
```
